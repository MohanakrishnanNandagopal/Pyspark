```python
#range funtion in python - to generate series of number...
#syntax - range(start,stop,step)
#default - start -- 0 step --1 stop must be given by user..
```

```python
#example 
a = range(0,5,1)
print(a[1])
```

```python
for i in a:
    print(i)
```

```python
a = range(-10,-1,1)
for i in a:
    print(i)
```

```python
#sum of even number between 1 and 100 which include 1 and 100
number = range(0,101,2)
sum = 0
for i in number:
    sum = sum + i
print(sum)
```

```python
number = range(1,101)
sum =0
for i in number:
    if(i % 2 == 0):
        sum = sum + i
print(sum)
    
```

```python
# Excercise 3 fizzbuzz form 1 to 100 3 - fizz 5 - buzz 15 - fizzbuzz
number = range(1,101)
for i in number:
    if (i % 3 ==0):
        print('fizz')
    if(i % 5 == 0):
        print('buzz')
    if(i % 3 == 0 and i % 5 ==0):
        print('fizzbuzz')
    else:
        print(i)
```

```python
number = range(1,101)
for i in number:
    if(i % 3 == 0 and i % 5 ==0):
        print('fizzbuzz')
    elif (i % 3 ==0):
        print('fizz')
    elif(i % 5 == 0):
        print('buzz')
    else:
        print(i)
```
